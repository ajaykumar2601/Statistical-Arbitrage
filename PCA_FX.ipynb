```python

import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Data preprocessing

import pandas as pd
base = '/Users/rana/Desktop/Ajay Kumar/Term 3/ZA/stats_arb_with excess returns/Boson'
file_path = base+'/boson data/G10 USD Crosses 1998-2023_1col.xlsx'
final_df = pd.DataFrame()
xls = pd.ExcelFile(file_path)

for sheet_name in xls.sheet_names:
    FX_data = pd.read_excel(xls, sheet_name=sheet_name, usecols='A,B', index_col=0)
    FX_data.columns = [sheet_name]
    if final_df.empty:
        final_df = FX_data.copy()
    else:
        final_df = final_df.merge(FX_data, left_index=True, right_index=True, how='outer')
final_df.fillna(method='ffill', inplace=True)

# for remaining nans backfill
final_df.fillna(method='bfill', inplace=True)

final_df_price = final_df[final_df.index <= '2022-12-31']

final_df_returns = final_df_price.pct_change()
final_df_returns = final_df_returns.rename_axis('Date')
final_df_returns.dropna(inplace=True)
display(final_df_returns)
```

```python
# !pip install sklearn
# !pip install scikit-learn

sfx_ret = final_df_returns
# Load the data
ff_data = pd.read_csv(base+'/boson data/F-F_Research_Data_5_Factors_2x3_daily.csv',index_col='Date')[['RF']]
ff_data.index = pd.to_datetime(ff_data.index)

sfx_ret = sfx_ret.merge(ff_data,how='left',on='Date')


sfx_ret['RF'].fillna(method='ffill',inplace=True)
sfx_ret = sfx_ret.applymap(pd.to_numeric, errors='coerce')

for _ in list(sfx_ret.columns)[:-1]:
    sfx_ret[_] = sfx_ret[_] - sfx_ret['RF']/100
    
sfx_ret.drop(columns=['RF'],inplace= True)
sfx_ret.to_csv(base+'/boson data/FX_Excess_Returns_final.csv')
print (sfx_ret.head(5))

window = 60
resid = np.empty(shape=sfx_ret.shape)
resid[:] = np.NAN
resid3 = np.empty(shape=sfx_ret.shape)
resid3[:] = np.NAN
resid5 = np.empty(shape=sfx_ret.shape)
resid5[:] = np.NAN

# Loop over each rolling window
for j in range(window, sfx_ret.shape[0]):
    try:
        # Extract the rolling window data from 0 to 60 (not including 60th index)
        rolling_ret = sfx_ret.iloc[(j - window):j, :]
        
        # Standardize the data
        scaler = StandardScaler()
        rolling_ret_scaled = scaler.fit_transform(rolling_ret)
        
        # Fit PCA 1-Factor ----------------------------------------------------
        pca = PCA(n_components=1)
        pca.fit(rolling_ret_scaled)
        
        # Standardize the asset returns at time j and project them onto the principal component (take 59th index)
        asset_ret_scaled = scaler.transform(sfx_ret.iloc[j-1:j, :])
        projection = pca.transform(asset_ret_scaled)
        residual = asset_ret_scaled - pca.inverse_transform(projection)
        
        # Store the residuals at 60th position
        resid[j-1, :] = residual
        

        # Fit PCA 3-Factor ----------------------------------------------------
        pca3 = PCA(n_components=3)
        pca3.fit(rolling_ret_scaled)
        
        # Standardize the asset returns at time j and project them onto the principal component
        asset_ret_scaled3 = scaler.transform(sfx_ret.iloc[j-1:j, :])
        projection3 = pca3.transform(asset_ret_scaled3)
        residual3 = asset_ret_scaled3 - pca3.inverse_transform(projection3)
        
        # Store the residuals
        resid3[j-1, :] = residual3

        # Fit PCA 3-Factor ----------------------------------------------------
        pca5 = PCA(n_components=3)
        pca5.fit(rolling_ret_scaled)
        
        # Standardize the asset returns at time j and project them onto the principal component
        asset_ret_scaled5 = scaler.transform(sfx_ret.iloc[j-1:j, :])
        projection5 = pca5.transform(asset_ret_scaled5)
        residual5 = asset_ret_scaled5 - pca5.inverse_transform(projection5)
        
        # Store the residuals
        resid5[j-1, :] = residual5
        
    except Exception as e:
        print("ERROR", j, str(e))
        resid[j-1, :] = np.nan
        resid3[j-1, :] = np.nan
        resid5[j-1, :] = np.nan
        


# Create a DataFrame for the residuals
resid_df = pd.DataFrame(resid, index=sfx_ret.index, columns=sfx_ret.columns)
resid_df.to_csv(base+'/boson data/FX_PCA_1_FACTOR_RESID_DAILY.CSV')

resid3_df = pd.DataFrame(resid3, index=sfx_ret.index, columns=sfx_ret.columns)
resid3_df.to_csv(base+'/boson data/FX_PCA_3_FACTOR_RESID_DAILY.CSV')

resid5_df = pd.DataFrame(resid5, index=sfx_ret.index, columns=sfx_ret.columns)
resid5_df.to_csv(base+'/boson data/FX_PCA_5_FACTOR_RESID_DAILY.CSV')
```
